```python
import pandas as pd

# Load datasets
df1 = pd.read_csv("heart.csv")
df2 = pd.read_csv("heart2.csv")

print("Shape DF1:", df1.shape)
print("Shape DF2:", df2.shape)

# Check column alignment
print("Columns DF1:", df1.columns.tolist())
print("Columns DF2:", df2.columns.tolist())

# Merge (stack) the datasets
df_master = pd.concat([df1, df2], axis=0, ignore_index=True)

print("Final merged shape:", df_master.shape)

# Optional: remove duplicates, if any
#df_master = df_master.drop_duplicates()
#print("Shape after removing duplicates:", df_master.shape)

# Save as new master CSV
df_master.to_csv("heart_master_raw.csv", index=False)

print("Saved new file: heart_master_raw.csv")
```
